```
#PLOTTING
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df.dropna(inplace=True)

#ploting a histogram using matplotlib and seaborn

plt.figure(figsize = (12,7))
sns.distplot(df['Respondent Age'], color='maroon')
plt.title('Distribution for Respondent Age')
plt.xlabel('Respondent age')
plt.ylabel('Frequency')

plt.show()
```

```
df['Respondent Age'].skew()
```

```
plt.figure(figsize = (8,5))
sns.countplot(df['country'])
plt.title('Country bar Chart',color ='red',fontdict={'size':15})
plt.xlabel
plt.show()
```

```
plt.figure(figsize = (12,8))
sns.countplot(df['marital_status'])
plt.title('MaritalStatus')
plt.show()
```
